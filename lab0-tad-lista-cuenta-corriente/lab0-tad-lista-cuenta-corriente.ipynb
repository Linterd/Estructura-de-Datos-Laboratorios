## 5. Tu tarea: operar sobre la lista

Completar el siguiente código para:

1. agregar un nuevo movimiento
2. calcular el saldo final
3. contar cuántos cargos y cuántos abonos existen

> Sugerencia:  
> - si el tipo es `"abono"`, el monto suma al saldo  
> - si el tipo es `"cargo"`, el monto resta al saldo

```python
# 1) Agrega un nuevo movimiento
movimientos.append({"descripcion": "Pago servicio de luz", "monto": 60000, "tipo": "cargo"})
```

```python
# 2) Calcula el saldo final
saldo = 0
for mov in movimientos:
    if mov["tipo"] == "abono":
        saldo += mov["monto"]
    elif mov["tipo"] == "cargo":
        saldo -= mov["monto"]
print("Saldo final: ", saldo, " $")

```

```python
# 3) Cuenta cuántos cargos y cuántos abonos existen
contar_cargos = 0
contar_abonos = 0

for mov in movimientos:
    if mov["tipo"] == "cargo":
        contar_cargos += 1
    elif mov["tipo"] == "abono":
        contar_abonos += 1
        
print("Cantidad de cargos:", contar_cargos)
print("Cantidad de abonos:", contar_abonos)
```

## 8. Tu tarea: completar la implementación

Se debe implementar los siguientes métodos:

- `obtener(posicion)`
- `eliminar(posicion)`
- `mostrar_todos()`
- `calcular_saldo()`

### Recomendaciones
- validar si la posición existe
- decidir cuándo conviene usar `return`
- probar cada método apenas se complete

```python
class ListaMovimientos:
    def __init__(self):
        self.movimientos = []

    def esta_vacia(self):
        return len(self.movimientos) == 0

    def agregar(self, movimiento):
        self.movimientos.append(movimiento)

    def tamano(self):
        return len(self.movimientos)

    def obtener(self, posicion):
        if 0 <= posicion < len(self.movimientos):
            return self.movimientos[posicion]
        
    def eliminar(self, posicion):
        if 0 <= posicion < len(self.movimientos):
            return self.movimientos.pop(posicion)
        else:
            return None
        
    def mostrar_todos(self):
        for mov in self.movimientos:
            print(mov)
            
    def calcular_saldo(self):
        saldo = 0
        for mov in self.movimientos:
            if mov["tipo"] == "abono":
                saldo += mov["monto"]
            elif mov["tipo"] == "cargo":
                saldo -= mov["monto"]
        return saldo
```
